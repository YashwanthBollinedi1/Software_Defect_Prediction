```python
# Import required packages
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# read data from csv using pandas and get the total number of null values present in the dataset.
data = pd.read_csv("jm1_csv.csv")
print("Total null values present:",data.isnull().sum().sum())
```

```python
#get the total number of null values for each column 
data.isnull().sum()
```

```python
# replace the null value with '0' and show the results
data1=data.fillna(0)
data1.isnull().sum()
```

```python
# read the column 'defects' and count number of 'True' and 'False' and plot the bar plot.
true_false_counts = data1['defects'].value_counts()
print(true_false_counts)
fig, ax = plt.subplots(figsize=(6, 4))
true_false_counts.plot(kind='bar', color=['blue', 'green'], ax=ax)
ax.set_title('Counts of Defects (True/False)', fontweight='bold')
ax.set_xlabel('Defects', fontweight='bold')
ax.set_ylabel('Count', fontweight='bold')
ax.set_xticklabels(['False', 'True'], rotation=0)
plt.show()
```

```python
# plot the correlation plot for dataset and the plot is saved as png file in project folder.
f,ax = plt.subplots(figsize = (15, 15))
ax=sns.heatmap(data1.corr(), cmap='RdYlGn_r', linewidths=0.5, annot=True)#annot 'true' shows the value in each cell.
plt.savefig("correlation.png")
```
